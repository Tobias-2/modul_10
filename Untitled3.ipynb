```python
def gauss(x, sigma=1, mu=3):
    return stats.norm.pdf(x, mu, sigma)
x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(12, 8))
plt.plot(x, gauss(x, sigma=1), label='odchylenie standardowe = 1, średnia = 3')
plt.legend(loc='upper left')
plt.show()
```

```python
data_100 = np.random.normal(loc=3, scale=1, size=100)
data_10000 = np.random.normal(loc=3, scale=1, size=10000)
```

```python
stats_100 = {
    'Średnia': np.mean(data_100),
    'Mediana': np.median(data_100),
    'Wariancja': np.var(data_100),
    'Odchylenie standardowe': np.std(data_100),
    'Minimum': np.min(data_100),
    'Maximum': np.max(data_100)
}

stats_10000 = {
    'Średnia': np.mean(data_10000),
    'Mediana': np.median(data_10000),
    'Wariancja': np.var(data_10000),
    'Odchylenie standardowe': np.std(data_10000),
    'Minimum': np.min(data_10000),
    'Maximum': np.max(data_10000)
}
```

```python
df_stats_100 = pd.DataFrame(stats_100, index=['Zbiór 100'])
df_stats_10000 = pd.DataFrame(stats_10000, index=['Zbiór 10,000'])
df_stats = pd.concat([df_stats_100, df_stats_10000])
```
